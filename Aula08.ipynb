crie um script Python simples para verificar a temperatura de um sistema. Se for maior que 30, exiba um alerta de superaquecimento.

```python
'''
atividade 1 - sistema de alerta de temperatura
'''
def check_system_temperature():
    # Solicita a temperatura ao usuário
    while True:
        try:
            user_input = input("Por favor, digite a temperatura atual do sistema em °C: ")
            current_temperature = float(user_input)
            break
        except ValueError:
            print("Entrada inválida. Por favor, digite um número para a temperatura.")

    # Demonstrando concatenação de strings usando o operador '+'
    # É necessário converter o número para string com str() antes de concatenar com '+'
    print("Temperatura atual do sistema: " + str(round(current_temperature, 2)) + "°C")

    if current_temperature > 30:
        # Exemplo de f-string, que é uma forma moderna de concatenação/interpolação
        print(f"ALERTA: Superaquecimento! A temperatura está acima de {current_temperature:.2f}°C.")
    else:
        print("A temperatura do sistema está normal.")

# Chama a função para verificar a temperatura
check_system_temperature()
```

```python
'''
atividade 2 - Organizar arquivos por extensão
'''
import os
import shutil
import random

def organize_files_by_extension(source_directory):
    if not os.path.isdir(source_directory):
        print(f"Erro: O diretório de origem '{source_directory}' não existe.")
        return

    print(f"Organizando arquivos em: {source_directory}")

    # Lista todos os itens no diretório de origem
    items = os.listdir(source_directory)

    for item in items:
        item_path = os.path.join(source_directory, item)

        # Ignora diretórios (a verificação '__file__' não é necessária em ambientes como Colab)
        if os.path.isdir(item_path):
            continue

        # Obtém a extensão do arquivo
        # os.path.splitext retorna uma tupla: ('nome_do_arquivo', '.extensao')
        _, extension = os.path.splitext(item)

        if extension:
            # Remove o ponto e converte para minúsculas (ex: '.JPG' -> 'jpg')
            extension = extension[1:].lower()
            destination_folder = os.path.join(source_directory, extension + '_files')

            # Cria a pasta de destino se ela não existir
            if not os.path.exists(destination_folder):
                os.makedirs(destination_folder)
                print(f"Pasta criada: {destination_folder}")

            # Move o arquivo para a pasta de destino
            try:
                shutil.move(item_path, destination_folder)
                print(f"Movido: '{item}' para '{destination_folder}'")
            except shutil.Error as e:
                print(f"Erro ao mover '{item}': {e}")
            except Exception as e:
                print(f"Ocorreu um erro inesperado ao mover '{item}': {e}")
        else:
            # Arquivos sem extensão podem ser movidos para uma pasta 'no_extension_files'
            no_extension_folder = os.path.join(source_directory, 'no_extension_files')
            if not os.path.exists(no_extension_folder):
                os.makedirs(no_extension_folder)
                print(f"Pasta criada: {no_extension_folder}")
            try:
                shutil.move(item_path, no_extension_folder)
                print(f"Movido: '{item}' para '{no_extension_folder}'")
            except shutil.Error as e:
                print(f"Erro ao mover '{item}': {e}")

    print("Organização de arquivos concluída!")

# --- Exemplo de uso ---
# Defina o diretório onde você quer organizar os arquivos.
# Se for o diretório atual, use '.'
# Para o diretório de conteúdo do Colab, use '/content'
source_directory = '/content' # Altere conforme necessário

# Certifica-se de que o diretório de origem existe
if not os.path.exists(source_directory):
    os.makedirs(source_directory)

# --- Cria 15 arquivos para teste ---
print("Criando 15 arquivos de teste...")
extensions = ['.pdf', '.jpg', '.png', '.txt', '.docx', '.xlsx', '.py', '.zip', ''] # '' para arquivos sem extensão
for i in range(1, 16):
    ext = random.choice(extensions)
    filename = f"arquivo_teste_{i}{ext}"
    filepath = os.path.join(source_directory, filename)
    try:
        with open(filepath, 'w') as f:
            f.write(f"Conteúdo do {filename}")
        print(f"Criado: {filename}")
    except Exception as e:
        print(f"Erro ao criar '{filename}': {e}")

print("Arquivos de teste criados. Iniciando a organização...")
organize_files_by_extension(source_directory)
```

```python
import time
import random

# Funções auxiliares para o programa autônomo de alerta
alert_count = 0 # Contador para controlar o número de alertas

def verificar_condicao():
    """
    Simula a verificação de uma condição crítica (ex: temperatura, uso de CPU).
    Retorna uma tupla (True se a condição for crítica, False caso contrário, valor_da_condicao).
    """
    # Para demonstração, vamos simular que a condição é crítica 50% das vezes
    # e variar o valor para mostrar diferentes mensagens
    if random.choice([True, False]):
        # Condição crítica
        valor = random.uniform(31.0, 45.0) # Temperatura acima do limite
        return True, valor
    else:
        # Condição normal
        valor = random.uniform(20.0, 30.0) # Temperatura dentro do normal
        return False, valor

def enviar_alerta_console(mensagem):
    """
    Exibe uma mensagem de alerta no console.
    """
    print(f"[ALERTA NO CONSOLE] {mensagem}")

def enviar_alerta_email_simulado(destinatario, assunto, corpo):
    """
    Simula o envio de um e-mail de alerta, imprimindo os detalhes no console.
    """
    print("\n--- E-MAIL SIMULADO ENVIADO ---")
    print(f"Para: {destinatario}")
    print(f"Assunto: {assunto}")
    print("Corpo do E-mail:")
    print(corpo)
    print("------------------------------\n")

def monitorar_condicao_critica():
    """
    Função principal que monitora continuamente uma condição e dispara alertas.
    """
    global alert_count
    print("Iniciando monitoramento de condições críticas...")
    print("Pressione 'Interromper execução' (quadrado) para parar.")

    # Loop de monitoramento (para demonstração, vamos rodar algumas vezes)
    for i in range(1, 6): # Monitora 5 vezes para demonstrar cenários
        print(f"\n--- Verificação {i}/5 ---")
        is_critical, valor_condicao = verificar_condicao()

        if is_critical:
            mensagem_alerta = f"Condição CRÍTICA detectada! Valor atual: {valor_condicao:.2f}"
            enviar_alerta_console(mensagem_alerta)

            # Enviar e-mail simulado apenas na primeira vez que a condição crítica for atingida
            # ou para demonstrar que o alerta é disparado
            if alert_count == 0 or random.choice([True, False]): # Simula que nem todo alerta crítico envia e-mail, ou envia o primeiro
                assunto_email = f"ALERTA CRÍTICO: {mensagem_alerta}"
                corpo_email = (
                    f"Prezado administrador,\n\n"
                    f"Uma condição crítica foi detectada no sistema.\n"
                    f"Detalhes: {mensagem_alerta}\n"
                    f"Ação recomendada: Investigar imediatamente.\n\n"
                    f"Este é um alerta automatizado, por favor não responda."
                )
                enviar_alerta_email_simulado("admin@example.com", assunto_email, corpo_email)
                alert_count += 1
        else:
            print(f"Condição NORMAL. Valor atual: {valor_condicao:.2f}")
            alert_count = 0 # Reseta o contador se a condição voltar ao normal

        time.sleep(2) # Espera 2 segundos antes da próxima verificação

    print("Monitoramento de demonstração concluído.")

# Inicia o monitoramento
monitorar_condicao_critica()
```
